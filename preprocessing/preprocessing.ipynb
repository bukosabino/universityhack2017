```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
print("Comenzando el preprocesamiento de datos . . . . . ")
print("Esta operación puede tardar algunos minutos . . . . . ")
```

```python
df = pd.read_csv('../data/train2.txt', sep="|").sort_values(by=["ID_Customer", "Cod_Fecha",])
```

```python
# Lista con cantidad de productos de cada cliente, ordenada.
n_prods = df.groupby("ID_Customer")["Cod_Prod"].count().values

# Lista con todos los productos en el mismo orden que aparecen
prods = df["Cod_Prod"].values

count = 0
# Diccionario con IDs como clave, y lista con todos sus productos ordenador como valor
productos = {}
ids = df.ID_Customer.unique()
for i in range(len(n_prods)):
    id = ids[i]
    # Cortamos prods a trozos, desde count hasta count + su cantidad
    productos[id] = list(prods[count:count+n_prods[i]])
    # Actualizamos count para avanzar
    count += n_prods[i]

# Binarizamos la columna "Cod_Prod"
bnrz = MultiLabelBinarizer()
labels_list = [productos[i][0:j] for i in ids for j in range(len(productos[i]))]
binarized = bnrz.fit_transform(labels_list)
del productos

# Añadimos las columnas Socio_Demo y las 94 columnas binarizadas
salida = pd.DataFrame()
salida["ID"] = df.ID_Customer
salida["SD1"] = df.Socio_Demo_01
salida["SD2"] = df.Socio_Demo_02
salida["SD3"] = df.Socio_Demo_03
salida["SD4"] = df.Socio_Demo_04
salida["SD5"] = df.Socio_Demo_05

for i in range(94):
    salida["P%i" % i] = binarized[:,i]

# El último producto adquirido se convierte en el target a predecir
salida["y"] = df.Cod_Prod
del df
del binarized
print "Dataset preprocesado"
```

```python
# Para la validación usamos directamente la última operación de cada cliente
val = salida.groupby("ID").last()
val['ID'] = salida['ID'].unique().tolist()
cols = val.columns.tolist()
cols.insert(0, cols.pop(-1))
val = val[cols]

# Generamos un índice con la última operación de cada cliente
ind = salida.reset_index().groupby('ID').index.last()

# Borramos de train los datos que usaremos para validar
train = salida.drop(ind)

print "Train/Validación separados"
```

```python
# Salidas de ficheros para pruebas externas
print("Preprocesamiento terminado. Guardando en archivos externos . . . . . ")
#train.to_hdf("../data/train_final.h5", "train", index=False)
#val.to_hdf("../data/validacion_final.h5", "val", index=False)
train.to_csv("../data/train_final.csv", sep=',', index=False)
val.to_csv("../data/validacion_final.csv", sep=',', index=False)
```
